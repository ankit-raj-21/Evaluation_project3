```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex']=df['Sex'].map({'M':3,'I':1,'F':2})
```

```python
df.head()
```

```python
df_abalone=df
df_abalone.shape
```

```python
for x in df_abalone.columns:
    if df_abalone.skew().loc[x]>0.55:
        df_abalone[x]=np.log1p(df_abalone[x])
```

```python
df_abalone.skew()
```

```python
x=df_abalone.drop(columns='Rings',axis=1)
y=df_abalone['Rings']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn=MinMaxScaler()
```

```python
x_scaled=mn.fit_transform(x)
```

```python
x_scaled
```

```python
x=pd.DataFrame(data=x_scaled,index=None,columns=x.columns)
```

```python
x.skew()
```

```python
x.shape
```

```python
x_columns=['Sex' 'Length','Diameter','Height','Wholeweight','Suckedweight','Visceraweight','Shellweight']
```

```python
y=pd.DataFrame(y)
```

```python
y.shape
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y, random_state=320,test_size=.33)
```

```python
lr=LinearRegression()
```

```python
lr.fit(train_x,train_y)
```

```python
lr.score(train_x,train_y)
```

```python
predict=lr.predict(test_x)
```

```python
r2_scor=r2_score(test_y,predict)
```

```python
r2_scor*100
```

```python
print('MSE=', mean_squared_error(test_y,predict)*100)
```

```python
print(lr.intercept_)
```

```python
print(lr.coef_)
```

```python
print(zip(x_columns, lr.coef_))
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNN
```

```python
kl=KNN(n_neighbors=7)
```

```python
kl.fit(train_x,train_y)
```

```python
kl.score(train_x,train_y)
```

```python
predi=kl.predict(test_x)
```

```python
r2kl=r2_score(test_y,predi)
```

```python
r2kl*100
```

```python
print("MSE=" , mean_squared_error(test_y,predi)*100)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dl=DecisionTreeRegressor(max_depth=12)
```

```python
dl.fit(train_x,train_y)
```

```python
dl.score(train_x,train_y)
```

```python
pred=dl.predict(test_x)
```

```python
r2dl_scor=r2_score(test_y,pred)
```
